```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dask.dataframe as dd
import time
```

```python
data_dir = '../Data/LANL_Earthquake_prediction/'
train_dir = data_dir + 'trainset/train.csv'
```

# Dask vs Pandas

## Dask

```python
print('Reading CSV with Dask')
t0 = time.time()
df = dd.read_csv(train_dir)
t1 = time.time()
print('time: %.4f s' % (t1-t0))
```

```python
print('Counting numer of rows with Dask')
t0 = time.time()
num_total = df.acoustic_data.count().compute()
t1 = time.time()
print('time: %.4f s' % (t1-t0))
```

## Pandas

```python
print('Reading CSV with Pandas')
t0 = time.time()
df = pd.read_csv(train_dir)
t1 = time.time()
print('time: %.4f s' % (t1-t0))
```

```python
print('Counting numer of rows with Pandas')
t0 = time.time()
num_total = df.acoustic_data.count()
t1 = time.time()
print('time: %.4f s' % (t1-t0))
```

# Plotting train data

```python
sampling_rate = 1000
subsampled = df.iloc[::sampling_rate, :]
num_rows = subsampled.acoustic_data.count()
print('Number of all samples: %d' % num_total)
print('Number of subsamples: %d' % num_rows)
```

```python
sample_num = np.arange(num_rows)
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Sample # (x%d)'%sampling_rate)
ax1.set_ylabel('acoustic data', color=color)  # we already handled the x-label with ax1
ax1.plot(sample_num, subsampled.acoustic_data, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:green'
ax2.set_ylabel('time to failure (s)', color=color)
ax2.plot(sample_num, subsampled.time_to_failure, color=color)
ax2.tick_params(axis='y', labelcolor=color)


plt.gcf().set_size_inches(15, 8)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

# Number of earthquakes and when they occur

```python
df['dtime'] = df['time_to_failure'] - df['time_to_failure'].shift(1)
segments = df.loc[df['dtime'] > 3]
```

```python
print('Number of earthquakes: %d' % len(segments))
print('Information on where new segments start:')
print(segments)
```
