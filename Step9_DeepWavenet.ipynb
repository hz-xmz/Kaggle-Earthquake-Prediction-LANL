# 2. Build Model

```python
from keras.utils import Sequence
from DataPreparation.dataset_preparation import uniform_from_slices, sample_sequences

seq_len = 150000
earthquake_ind = [0, 5656574, 50085878, 104677356,
                  138772453, 187641820, 218652630, 245829585,
                  307838917, 338276287, 375377848, 419368880,
                  461811623, 495800225, 528777115, 585568144,
                  621985673]
train_eq_ind = earthquake_ind[:-4]
train_slicing_ranges = [(train_eq_ind[i], (train_eq_ind[i+1]-seq_len)) for i in range(len(train_eq_ind)-1)]
train_sample_size = n # 3355

len_train_total = len(X_train) + len(X_val)
val_eq_ind = np.hstack([earthquake_ind[-4:], len_train_total])
val_eq_ind -= np.min(val_eq_ind)
val_slicing_ranges = [(val_eq_ind[i], (val_eq_ind[i+1]-seq_len)) for i in range(len(val_eq_ind)-1)]

class MY_Generator(Sequence):
    def __init__(self, inputs, labels, batch_size, num_samples, slicing_ranges):
        self.inputs, self.labels = inputs, labels
        self.batch_size, self.num_samples = batch_size, num_samples
        self.slicing_ranges = slicing_ranges

    def __len__(self):
#         return np.ceil(len(self.inputs) / float(self.batch_size))
        return int(float(self.num_samples) / float(self.batch_size))

    def __getitem__(self, idx):
#         batch_x = self.image_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]
#         batch_y = self.labels[idx * self.batch_size:(idx + 1) * self.batch_size]        
        sample_ind = uniform_from_slices(self.slicing_ranges, self.batch_size)
        batch_x, batch_y = sample_sequences(self.inputs, self.labels, sample_ind, seq_len)
        batch_x = batch_x[:, :, None]
        return batch_x, batch_y
batch_size = 16
num_train_samples = 10000
num_validation_samples = 5000
train_generator = MY_Generator(X_train, y_train, batch_size, num_train_samples, train_slicing_ranges)
val_generator = MY_Generator(X_val, y_val, batch_size, num_validation_samples, val_slicing_ranges)
```

```python
def dilation_layer(dilation_channels, residual_channels, skip_channels, filter_width, dilation_rate, name):
    def f(input_batch):
        conv_filter = keras.layers.Conv1D(dilation_channels, filter_width,
                                       strides=1, dilation_rate=dilation_rate,
                                       padding='same', use_bias=True, 
                                       activation='tanh', name='filter_'+name)(input_batch)
        conv_gate = keras.layers.Conv1D(dilation_channels, filter_width,
                                       strides=1, dilation_rate=dilation_rate,
                                       padding='same', use_bias=True, 
                                       activation='sigmoid', name='gate_'+name)(input_batch)
        out = keras.layers.Multiply(name='out_'+name)([conv_filter, conv_gate])

        transformed = keras.layers.Conv1D(residual_channels, 1, 
                                          padding='same', use_bias=True, 
                                          activation='linear', name='transformed_'+name)(out)
        skip_contribution = keras.layers.Conv1D(skip_channels, 1,
                                                padding='same', use_bias=True,
                                                activation='linear', name='skip_'+name)(out)
        
        return keras.layers.Add(name='resid_'+name)([transformed, input_batch]), skip_contribution  
    return f

residual_channels = 128
dilation_channels = 32
skip_channels = 512
filter_width = 2
num_outputs = 10
initial_filter_width = 32
dilations = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512] * 5

def WaveNet(input_shape=(seq_len, 1), dilations=dilations):    
    input_x = keras.layers.Input(shape=input_shape, name='input')
    
    x = keras.layers.Conv1D(128, 100,
                           strides = 100,
                           padding='valid', use_bias=True, 
                           activation='tanh')(input_x)
    
    out_list = []
    for i, dilation in enumerate(dilations):
        x, out_ = dilation_layer(dilation_channels, residual_channels, skip_channels, filter_width, dilation, str(i))(x)
        out_list.append(out_)
    
    skip_overall = keras.layers.Add(name='skip_overall')(out_list)
    skip_relu = keras.layers.ReLU(name='skip_relu')(skip_overall)
    
    postprocess_conv = keras.layers.Conv1D(skip_channels, 1, padding='same',
                                           activation='relu', name='postprocess_conv1')(skip_relu)
    
    postprocess_conv = keras.layers.Conv1D(128, 1, padding='same',
                                           activation='linear', name='postprocess_conv2')(postprocess_conv)

    out = keras.layers.Flatten()(postprocess_conv)
    out = keras.layers.Dense(1, kernel_initializer='normal',activation='linear')(out)
    
    return keras.models.Model(inputs=input_x, outputs=out)
```

```python
model = WaveNet()
earlyStopping = EarlyStopping(monitor='val_loss',
                              patience=15,
                              verbose=0,
                              mode='min')
mcp_save = ModelCheckpoint('.mdl_wts.hdf5',
                           save_best_only=True,
                           monitor='val_loss',
                           mode='min')
reduce_lr_loss = ReduceLROnPlateau(monitor='val_loss',
                                   factor=0.1,
                                   patience=5,
                                   verbose=1,
                                   epsilon=1e-4,
                                   mode='min')

model.compile(loss='mean_absolute_error', optimizer= adam(lr=1e-4, decay=1e-4), metrics=['mean_absolute_error'])
model.summary()
```

```python
mae_diff = np.abs(y_val_-y_pred)
top5_best_ind = mae_diff.argsort()[:10]
top5_best = X_val_[top5_best_ind]
top5_best_target = y_val_[top5_best_ind]
top5_best_pred = y_pred[top5_best_ind]

top5_worst_ind = mae_diff.argsort()[-10:][::-1]
top5_worst = X_val_[top5_worst_ind]
top5_worst_target = y_val_[top5_worst_ind]
top5_worst_pred = y_pred[top5_worst_ind]
```

```python
sampling_rate = 1
subsampled_best = top5_best[:, ::sampling_rate]
subsampled_worst = top5_worst[:, ::sampling_rate]
num_rows = top5_best.shape[1]
print(top5_best.shape)
print(subsampled_best.shape)
print('-----')
print(top5_worst.shape)
print(subsampled_worst.shape)
```

```python
mean_time_per_step = np.mean(np.asarray([y_val[i]-y_val[i+1] for i in range(100)]))
print(mean_time_per_step)
```

```python
time_to_failure = lambda x: np.linspace((mean_time_per_step * num_rows) + x, x, num_rows)
```
